```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("live_action_remakes.csv")
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 9))

for _, row in df.iterrows():
    ax.plot(["Animated", "Live Action"], [row["animated_rt"], row["live_action_rt"]],
            marker="o", label=row["Movie"])

# Formatting
ax.set_ylabel("Rotten Tomatoes Score")
ax.set_title("Comparison of Animated vs Live Action Ratings")
ax.set_xticks(["Animated", "Live Action"])  # Ensure correct x-axis labels

# Add legend
ax.legend(title="Movie", bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```
